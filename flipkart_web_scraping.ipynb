## Flipkart Gaming laptops - Web Scraping

### **Import Necessary Libraries**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

### **Read the url**

```python
url = requests.get("https://www.flipkart.com/6bo/b5g/~cs-6f2w95iiwq/pr?sid=6bo%2Cb5g&collection-tab-name=Gaming+Laptops&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIkdhbWluZyJdLCJ2YWx1ZVR5cGUiOiJNVUxUSV9WQUxVRUQifX19fX0%3D&wid=5.productCard.PMU_V2_4")

# Converting into beautiful soup object

soup = bs(url.content)
print(soup.prettify())
```

### **Function to get page URLs**

```python
url = soup.select(".WSL9JP a")
link = [a.get('href') for a in url]
```

```python
def get_page_url(base_url):

    page_url = []

    for urls in link:
        url_link = base_url+urls
        page_url.append(url_link)
    
    return page_url
```

### **Scrape laptop data**

```python
#Functin to scrape laptop data
def scrape_laptop_data(page_url): 
    try:
        res = requests.get(page_url)
        soup = bs(res.text,'html.parser')
    
        laptops = []
        for item in soup.find_all('div',class_="_75nlfW"):
            laptop = {}

            #get latop name
            name_tag = item.find('div',class_="KzDlHZ")
            if name_tag:
                laptop['name'] = name_tag.text.strip()

            #get specs
            specs = item.find('ul',class_="G4BRas")
            spec_list = [specs.text.strip() for spec in specs]
            laptop['specification'] = spec_list

            #get price
            price_tag = item.find('div',class_="Nx9bqj _4b5DiR")
            if price_tag:
                laptop['price'] = price_tag.text.strip()

            ratings = item.find('div', class_='XQDdHH')
            if ratings:
                laptop['ratings'] = ratings.text.strip()

            if laptop.get('name'):
                laptops.append(laptop)
            
        return laptops
    except Exception as e:
        print(f"Error scraping {page_url}: {e}")
        return[]
```

```python
#Generate page urls
base_url = "https://www.flipkart.com"

page_url = get_page_url(base_url)
    
# Exclude last url
page_url = page_url[:-1]
page_url
```

```python
# Scrape the data for all laptops
all_laptops =[]
for page_urls in page_url:
    laptops = scrape_laptop_data(page_urls)
    all_laptops.extend(laptops)
```

```python
# convert the list of dictionaries to a dataframe
laptop_df = pd.DataFrame(all_laptops)

laptop_df
```
